# Demo: Multimodal Agents Lab with Validation

This notebook demonstrates how the validation system integrates with the lab exercises.

## Step 1: Load the Validation System

```python
# First, we'll load the validation system
# In the real lab, this would be: %run validation_system.ipynb
# For this demo, I'll include the key components inline

import inspect
from typing import Any, Tuple, List, Dict, Callable
from datetime import datetime
from IPython.display import display, HTML
import ipywidgets as widgets

class ValidationResult:
    """Represents the result of a validation check"""
    def __init__(self, passed: bool, message: str, hints: List[str] = None, details: Dict = None):
        self.passed = passed
        self.message = message
        self.hints = hints or []
        self.details = details or {}
        self.timestamp = datetime.now()
    
    def display(self):
        """Display validation result with formatting"""
        if self.passed:
            status_html = '<div style="padding: 10px; background-color: #d4edda; border: 1px solid #c3e6cb; border-radius: 5px; margin: 10px 0;">'
            status_html += '<h3 style="color: #155724;">✅ Validation Passed!</h3>'
            status_html += f'<p style="color: #155724;">{self.message}</p>'
        else:
            status_html = '<div style="padding: 10px; background-color: #f8d7da; border: 1px solid #f5c6cb; border-radius: 5px; margin: 10px 0;">'
            status_html += '<h3 style="color: #721c24;">❌ Validation Failed</h3>'
            status_html += f'<p style="color: #721c24;">{self.message}</p>'
            
            if self.hints:
                status_html += '<h4 style="color: #856404;">💡 Hints:</h4>'
                status_html += '<ul style="color: #856404;">'
                for hint in self.hints:
                    status_html += f'<li>{hint}</li>'
                status_html += '</ul>'
        
        status_html += '</div>'
        display(HTML(status_html))

class LabProgress:
    """Tracks student progress through the lab"""
    def __init__(self):
        self.completed_blocks = set()
        self.attempts = {}
        self.start_time = datetime.now()
    
    def record_attempt(self, block_num: int, passed: bool):
        if block_num not in self.attempts:
            self.attempts[block_num] = []
        self.attempts[block_num].append({
            'passed': passed,
            'timestamp': datetime.now()
        })
        if passed:
            self.completed_blocks.add(block_num)
    
    def display_progress(self, total_blocks: int = 3):  # Using 3 for demo
        percentage = (len(self.completed_blocks) / total_blocks) * 100
        completed = len(self.completed_blocks)
        
        progress_html = f'''
        <div style="margin: 20px 0;">
            <h3>📊 Your Progress: {completed}/{total_blocks} CODE_BLOCKs completed ({percentage:.1f}%)</h3>
            <div style="width: 100%; background-color: #e0e0e0; border-radius: 10px; overflow: hidden;">
                <div style="width: {percentage}%; background-color: #4CAF50; height: 30px; text-align: center; line-height: 30px; color: white;">
                    {percentage:.1f}%
                </div>
            </div>
        </div>
        '''
        display(HTML(progress_html))

# Global progress tracker
lab_progress = LabProgress()

def create_validation_widget(block_num: int, validator_func: Callable, variable_name: str):
    """Creates an interactive validation widget for a CODE_BLOCK"""
    
    button = widgets.Button(
        description=f'Validate CODE_BLOCK_{block_num}',
        button_style='primary',
        icon='check'
    )
    
    output = widgets.Output()
    
    def on_click(b):
        with output:
            output.clear_output()
            
            try:
                var = globals().get(variable_name)
                if var is None:
                    result = ValidationResult(
                        False,
                        f"Variable '{variable_name}' not found.",
                        hints=[f"Make sure you've run the CODE_BLOCK_{block_num} cell and created '{variable_name}'"]
                    )
                else:
                    result = validator_func(var)
                    lab_progress.record_attempt(block_num, result.passed)
                
                result.display()
                
                if result.passed:
                    lab_progress.display_progress()
                    
            except Exception as e:
                result = ValidationResult(
                    False,
                    f"Error during validation: {str(e)}",
                    hints=["Check your code for syntax errors"]
                )
                result.display()
    
    button.on_click(on_click)
    
    return widgets.VBox([button, output])

print("✅ Validation system loaded successfully!")
```

## Exercise 1: Simple Variable Assignment

Let's start with a simple example. Create a variable called `my_name` with your name.

```python
# CODE_BLOCK_1: Replace this with your code
# Create a variable called 'my_name' with your name as a string

# <CODE_BLOCK_1>
# my_name = "Your Name Here"
```

```python
# Let's define a validator for this exercise
def validate_code_block_1(name_var):
    if not isinstance(name_var, str):
        return ValidationResult(
            False,
            "The variable should be a string.",
            hints=["Make sure to use quotes around your name", "Example: my_name = \"John Doe\""]
        )
    
    if len(name_var) == 0:
        return ValidationResult(
            False,
            "The name shouldn't be empty.",
            hints=["Add your actual name to the string"]
        )
    
    if name_var == "Your Name Here":
        return ValidationResult(
            False,
            "Please replace the placeholder with your actual name.",
            hints=["Change \"Your Name Here\" to your real name"]
        )
    
    return ValidationResult(
        True,
        f"Great job, {name_var}! You've completed your first CODE_BLOCK! 🎉"
    )

# Display the validation widget
display(create_validation_widget(1, validate_code_block_1, 'my_name'))
```

## Exercise 2: Create a Function

Now let's create a function that processes text. This simulates the PDF processing function from the real lab.

```python
# CODE_BLOCK_2: Create a function called 'process_text' that:
# - Takes a text parameter
# - Converts it to uppercase
# - Returns the result

# <CODE_BLOCK_2>
# def process_text(text):
#     return text.upper()
```

```python
# Validator for the function
def validate_code_block_2(func):
    if not callable(func):
        return ValidationResult(
            False,
            "process_text should be a function.",
            hints=["Make sure you defined it with 'def process_text(text):'"]    
        )
    
    # Test the function
    try:
        test_input = "hello world"
        result = func(test_input)
        
        if result != "HELLO WORLD":
            return ValidationResult(
                False,
                f"Function returned '{result}' but expected 'HELLO WORLD'.",
                hints=[
                    "Use the .upper() method on the text parameter",
                    "Make sure to return the result"
                ]
            )
        
        return ValidationResult(
            True,
            "Excellent! Your function works correctly. 🎯"
        )
        
    except Exception as e:
        return ValidationResult(
            False,
            f"Function raised an error: {str(e)}",
            hints=["Check your function implementation", "Make sure it accepts one parameter"]
        )

# Display validation widget
display(create_validation_widget(2, validate_code_block_2, 'process_text'))
```

## Exercise 3: Working with Lists (Simulating MongoDB insertion)

Create a list of dictionaries to simulate document insertion.

```python
# CODE_BLOCK_3: Create a list called 'documents' containing at least 3 dictionaries
# Each dictionary should have 'title' and 'content' keys

# <CODE_BLOCK_3>
# documents = [
#     {"title": "Document 1", "content": "This is the first document"},
#     {"title": "Document 2", "content": "This is the second document"},
#     {"title": "Document 3", "content": "This is the third document"}
# ]
```

```python
# Validator for documents
def validate_code_block_3(docs):
    if not isinstance(docs, list):
        return ValidationResult(
            False,
            "documents should be a list.",
            hints=["Create a list using square brackets: documents = [...]"]
        )
    
    if len(docs) < 3:
        return ValidationResult(
            False,
            f"You have {len(docs)} documents, but need at least 3.",
            hints=["Add more dictionary items to your list"]
        )
    
    for i, doc in enumerate(docs):
        if not isinstance(doc, dict):
            return ValidationResult(
                False,
                f"Item {i} is not a dictionary.",
                hints=["Each item should be a dictionary with curly braces: {...}"]
            )
        
        if 'title' not in doc or 'content' not in doc:
            return ValidationResult(
                False,
                f"Document {i} is missing required keys.",
                hints=["Each dictionary needs both 'title' and 'content' keys"]
            )
    
    return ValidationResult(
        True,
        f"Perfect! You created {len(docs)} valid documents. 📚"
    )

# Display validation widget
display(create_validation_widget(3, validate_code_block_3, 'documents'))
```
